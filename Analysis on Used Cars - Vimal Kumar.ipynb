```python
#replacing missing values of of drive column by mode
cars['drive'] = cars['drive'].fillna(cars['drive'].mode()[0])
```

```python
#replacing missing values of of engV column by median
cars['engV'] = cars['engV'].fillna(cars['engV'].median())
```

```python
cars.engV.isnull().sum()
```

```python
#replacing missing values of of price column by median using groupby
cars['price'].replace(0,np.nan, inplace=True)
cars['price'].fillna(cars.groupby(['car','body','engType'])['price'].transform('median'),inplace=True)
cars['price'].fillna(cars.groupby(['car','year'])['price'].transform('median'),inplace=True)
cars['price'].fillna(cars.groupby(['car'])['price'].transform('median'),inplace=True)
cars['price'].fillna(cars.groupby(['year'])['price'].transform('median'),inplace=True)
```

```python
cars['price'].isnull().sum()
```

```python
#replacing missing values of of price column by median using groupby
cars['mileage'].replace(0,np.nan, inplace=True)
```

```python
cars['mileage'].fillna(cars.groupby(['car','body','engType'])['mileage'].transform('median'),inplace=True)
cars['mileage'].fillna(cars.groupby(['car','body',])['mileage'].transform('median'),inplace=True)
```
